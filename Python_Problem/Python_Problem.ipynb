```python
import json

# Import the tweepy library.
# A tweepy library supports accessing the twitter.
import tweepy
```

```python
# Variables that contains the user credentials to access Twitter API.
# These variable can be accessed in twitter developers account in key and access token tab upon creating a new application. 
ACCESS_TOKEN = '..........'
ACCESS_SECRET = '..........'
CONSUMER_KEY = '..........'
CONSUMER_SECRET = '..........'
```

```python
# Setup tweepy to authenticate with Twitter credentials:
# This will create an OAuthHandler instance into which we pass our consumer token and secret.
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
```

```python
# Create the api to connect to twitter with your creadentials
api = tweepy.API(auth)
```

```python
# Here I have created a jsonl file and dump the data of home timeline of the twitter into the jsonl file.
# This snippet will take upto 200 tweets per page and I have taken tweets from total of 4 pages
with open('home_timeline_twitter_data.jsonl', 'w') as f:
    for page in tweepy.Cursor(api.home_timeline, count=200).pages(4):
        for status in page:
            f.write(json.dumps(status._json)+"\n")
```

```python
#  Here I have opened the jsonl file and loop through all the tweets which is in json format.
# I have retrieve the text, date & time, favorite count, retweet count, images of the tweet.
tweets_file = open('home_timeline_twitter_data.jsonl', "r")

for line in tweets_file:
        # Read in one line of the file, convert it into a json object 
        tweet = json.loads(line)
         
        if 'text' in tweet: # only messages contains 'text' field is a tweet
            print(tweet['text']) 
            print(tweet['created_at']) 
            print(tweet['favorite_count']) 
            print(tweet['retweet_count'])
            #print(tweet['profile_image_url'])
            try:
                print(True in [medium['type'] == 'photo' for medium in tweet.entities['media']])
            except:
                print("None")
            
```
