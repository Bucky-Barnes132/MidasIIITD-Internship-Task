```python
# Importing the necessary library for handling the data.
import pickle
import pandas as pd
```

```python
# Load the pickle 'train_image.pkl' file
with open(r'C:\Users\LENOVO\Desktop\train_image.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
# Load the pickle 'train_label.pkl' file
with open(r'C:\Users\LENOVO\Desktop\train_label.pkl', 'rb') as g:
    data_label = pickle.load(g)
```

```python
# Importing the train_test_split method for splitting the data into train and test.
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, data_label, test_size=0.3, random_state=8)
```

```python
# Converting the pickle data into the pandas dataframe
data = pd.DataFrame(data)
data_label = pd.DataFrame(data_label)
```

```python
xx = pd.DataFrame(X_train)
xx
```

```python
xt = pd.DataFrame(X_test)
xt
```

```python
yy = pd.DataFrame(y_train)
yy
```

```python
yt = pd.DataFrame(y_test)
yt
```

```python
# Using xgboost model for fitting the data to the model.
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(xx,yy)
```

```python
# The accuracy of the training data
xgb.score(xx,yy)
```

```python
# Predicting the accuracy for the test class.
pre = xgb.predict(xt)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(yt,pre)
score
```
